```python
#Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
#have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data

# import the required packages
import selenium 
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time

#connect to the web driver
driver=webdriver.Chrome(r"C:\Users\Asus i3\Downloads\chromedriver_win32\chromedriver.exe")

#Opening the Naukri page on automated web browser
driver.get("https://www.naukri.com/")

#Enter Designation and Location as required
des=driver.find_element(By.CLASS_NAME,"suggestor-input ")
des.send_keys('Data Analyst')

loc=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
loc.send_keys('Bangalore')

search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()

#scrap job title from given page
job_title=[]
tit=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in tit[0:10]:
    aa=i.text
    job_title.append(aa)
job_title

#scrap Job Location
loc=[]
lit=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in lit[0:10]:
    bb=i.text
    loc.append(bb)
loc

#scrap Company name
cname=[]
cc=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in cc[0:10]:
    cc=i.text
    cname.append(cc)
cname
#scrap Job Experience
exp=[]
je=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in je[0:10]:
    dd=i.text
    exp.append(dd)
exp

print(len(job_title),len(loc),len(cname),len(exp))
```

```python
#Q8 Write a python program to scrape data for Top 1000 Quotes of All Time.

#import the required libraries
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time

#connect to the web driver
driver=webdriver.Chrome(r"C:\Users\Asus i3\Downloads\chromedriver_win32\chromedriver.exe")

#First get the webpage https://www.azquotes.com/
driver.get("https://www.azquotes.com/")

#Click on Top Quotes
quotes=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
quotes.click()

#Than scrap a) Quote b) Author c) Type Of Quotes
quote_names=[]
for page in range(0,9):
    qn=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in qn:
        quote_names.append(i.text)
    next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
    next_button.click()
    time.sleep(5)

len(quote_names)

#scrape Author
author_names=[]
for page in range(0,9):
    an=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in an:
        author_names.append(i.text)
    next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
    next_button.click()
    time.sleep(5)
len(author_names)

#scrape type of quotes
type_names=[]
for page in range(0,9):
    ty=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in ty:
        type_names.append(i.text)
    next_button=driver.find_element(By.XPATH,'//li[@class="next"]')
    next_button.click()
    time.sleep(5)
len(author_names)

print(len(quote_names),len(author_names),len(type_names))
```

```python
print(len(quote_names),len(author_names),len(type_names))
df8=pd.DataFrame({'Quote':quote_names,'Author':author_names,'Type of quotes':type_names})
df8
```
