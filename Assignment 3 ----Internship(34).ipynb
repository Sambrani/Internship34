```python
#Q9 Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException
from selenium.webdriver.common.by import By
import time
import urllib

#connect to the web driver
driver=webdriver.Chrome(r"C:\Users\Asus i3\Downloads\chromedriver_win32.chromedriver.exe")

#connect to https://www.hostelworld.com/--------------------------------------------------------------------------
driver.get("https://www.youtube.com/watch?v=Xjv1sY630Uc&ab_channel=TechWithTim")
time.sleep(3)

#Scrape comments---------------------------------------------------------------------------------------------------
comm=[]
for _ in range(500):
    driver.execute_script("window.scrollBy(0,500)")
    try:
        com=driver.find_elements(By.CLASS_NAME,'style-scope ytd-comment-renderer')
        for i in com:
            comm.append(i.text)
    except NOSuchElementException:
        comm.append("-")
print(comm)

#Scrape -------------------------------------------------------------------
likes=[]
for _ in range(500):
    driver.execute_script("window.scrollBy(0,100)")
    try:
        li=driver.find_elements(By.CLASS_NAME,'style-scope ytd-comment-action-buttons-renderer')
        for i in li:
            likes.append(i.text)
    except NOSuchElementException:
        likes.append("-")
print(likes)
```

```python
#Q10 Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location.
#You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException
from selenium.webdriver.common.by import By
import time
import urllib

#connect to the web driver
driver=webdriver.Chrome(r"C:\Users\Asus i3\Downloads\chromedriver_win32.chromedriver.exe")

#connect to https://www.hostelworld.com/--------------------------------------------------------------------------
driver.get("https://www.hostelworld.com/")
time.sleep(3)

d_list1=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[1]/header/div/button[2]")
d_list1.click()
time.sleep(3)

d_list2=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[1]/header/div[2]/div/div[2]/ul[2]/li[2]/a")
d_list2.click()
time.sleep(3)

d_list3=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div/div[3]/ul/li[1]/div/ul/li[12]/a")
d_list3.click()
time.sleep(3)

d_list4=driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/div[1]/div[3]/div[1]/div[1]/div[2]/h2/a")
d_list4.click()
time.sleep(3)

#Scrape individual URL's-----------------------------------------------------------------------------------------------
urls=[]
for page in range(1,4):
    ur=driver.find_elements(By.XPATH,'//a[@class="view-button"]')
    for i in ur:
        urls.append(i.get_attribute('href'))
    nxt_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[4]/div/div[2]/section/div/div[7]')
    nxt_button.click()
    time.sleep(3)
print(urls)

#Scrape hostel name--------------------------------------------------------------------------------------------------
hname=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        hn=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/header/div[2]/div/div/h1')
        hname.append(hn.text)
    except NoSuchElementException:
        hname.append("-")
print(hname)

#Scrape distance from city centre------------------------------------------------------------------------------------
dist=[]
for page in range(1,4):
    di=driver.find_elements(By.XPATH,'//span[@class="description"]')
    for i in di:
        dist.append(i.text)
    nxt_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[4]/div/div[2]/section/div/div[7]')
    nxt_button.click()
    time.sleep(3)
print(dist)

#Scrape ratings------------------------------------------------------------------------------------------------------
ratings=[]
for i in urls:
    driver.get(i)
    time.sleep(2)
    try:
        rt=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[1]/div[3]/div/div[1]/a/div/div[1]')
        ratings.append(rt.text)
    except NoSuchElementException:
        ratings.append("-")
print(ratings)

#Scrape total reviews--------------------------------------------------------------------------------------------------
rev=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        rv=driver.find_element(By.CLASS_NAME,'reviews')
        rev.append(rv.text)
    except NoSuchElementException:
        rev.append("-")
print(rev)

#Scrape facilities-------------------------------------------------------------------------------------------------- 
facilities=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        fc=driver.find_element(By.CLASS_NAME,'facility-sections')
        facilities.append(fc.text)
    except NoSuchElementException:
        facilities.append("-")
print(facilities)
#Scrape property description--------------------------------------------------------------------------------------------------
descr=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        ds=driver.find_element(By.CLASS_NAME,'description-body')
        descr.append(ds.text)
    except NoSuchElementException:
        descr.append("-")
print(descr)
```

```python
print(len(hname),len(urls),len(dist),len(ratings),len(rev),len(facilities),len(descr))
```
